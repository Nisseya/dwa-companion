```python
import numpy as np

# Franklin, TN coordinates
franklin_lat, franklin_lon = 35.9251, -86.8689

# Convert degrees to radians
lat1 = np.radians(df["Lattitude"])
lon1 = np.radians(df["Longtitude"])
lat2 = np.radians(franklin_lat)
lon2 = np.radians(franklin_lon)

# Haversine formula
dlat = lat2 - lat1
dlon = lon2 - lon1
a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
c = 2 * np.arcsin(np.sqrt(a))
r = 6371  # Earth radius in km
df["Distance_km"] = r * c
```

```python
# images_to_fixed_pdf.py
# deps: pip install pillow reportlab

import os
from pathlib import Path
from PIL import Image, ImageOps
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader

# === CONFIGURATION ===
FOLDER = r"C:\Users\AIO\Downloads\Chapter page _ 3 books-20250918T184628Z-1-001\Chapter page _ 3 books"   # <--- set this
OUTPUT = Path(FOLDER) / "_pdf_5.5x8.5"

# Page size in points (1 in = 72 pt)
INCH = 72
PAGE_W, PAGE_H = int(5.5 * INCH), int(8.5 * INCH)
IMAGE_EXTS = {".jpg", ".jpeg", ".png", ".webp", ".bmp", ".tif", ".tiff"}


def fit_box(img_w, img_h, box_w, box_h):
    """Scale image to fit inside box while keeping aspect ratio."""
    if img_w == 0 or img_h == 0:
        return 0, 0
    scale = min(box_w / img_w, box_h / img_h)
    return int(img_w * scale), int(img_h * scale)


def image_to_pdf_page(img_path: Path, pdf_path: Path):
    pdf_path.parent.mkdir(parents=True, exist_ok=True)

    with Image.open(img_path) as im:
        im = ImageOps.exif_transpose(im)
        if im.mode not in ("RGB", "RGBA"):
            im = im.convert("RGB")
        if im.mode == "RGBA":
            bg = Image.new("RGB", im.size, "white")
            bg.paste(im, mask=im.split()[-1])
            im = bg

        img_w, img_h = im.size
        draw_w, draw_h = fit_box(img_w, img_h, PAGE_W, PAGE_H)
        x, y = (PAGE_W - draw_w) // 2, (PAGE_H - draw_h) // 2

        c = canvas.Canvas(str(pdf_path), pagesize=(PAGE_W, PAGE_H))
        c.setFillColorRGB(1, 1, 1)
        c.rect(0, 0, PAGE_W, PAGE_H, stroke=0, fill=1)
        c.drawImage(ImageReader(im), x, y, width=draw_w, height=draw_h, mask="auto")
        c.showPage()
        c.save()


def run():
    count = 0
    for root, _, files in os.walk(FOLDER):
        for name in files:
            if Path(name).suffix.lower() in IMAGE_EXTS:
                img_path = Path(root) / name
                rel = img_path.relative_to(FOLDER).with_suffix(".pdf")
                pdf_path = OUTPUT / rel
                try:
                    image_to_pdf_page(img_path, pdf_path)
                    count += 1
                    print(f"OK  -> {pdf_path}")
                except Exception as e:
                    print(f"FAIL-> {img_path} ({e})")
    print(f"\nDone. Converted {count} image(s) into PDFs at {OUTPUT}")


if __name__ == "__main__":
    run()
```

```python
# side_by_side_pdf.py
# deps: pip install pypdf reportlab

from pathlib import Path
from pypdf import PdfReader
from reportlab.pdfgen import canvas

# === CONFIGURATION ===
PDF_LEFT  = Path(r"C:\Users\AIO\Downloads\Chapter page _ 3 books-20250918T184628Z-1-001\Chapter page _ 3 books\BI_chapters+covers (5x5 in x 8x5 in)\BI_back_cover.png")
PDF_RIGHT = Path(r"c:\Users\AIO\Downloads\Chapter page _ 3 books-20250918T184628Z-1-001\Chapter page _ 3 books\BI_chapters+covers (5x5 in x 8x5 in)\BI_cover.png")
OUTPUT    = Path(r"C:\Users\AIO\Downloads\Chapter page _ 3 books-20250918T184628Z-1-001\Chapter page _ 3 books\BI_chapters+covers (5x5 in x 8x5 in)\BI_back_cover.png")


def merge_side_by_side(pdf1: Path, pdf2: Path, out: Path):
    r1, r2 = PdfReader(str(pdf1)), PdfReader(str(pdf2))

    if len(r1.pages) != len(r2.pages):
        raise ValueError("PDFs must have the same number of pages")

    writer_canvas = canvas.Canvas(str(out))

    for i in range(len(r1.pages)):
        p1 = r1.pages[i]
        p2 = r2.pages[i]

        w1, h1 = float(p1.mediabox.width), float(p1.mediabox.height)
        w2, h2 = float(p2.mediabox.width), float(p2.mediabox.height)

        # Output page size = max height, sum widths
        out_w = w1 + w2
        out_h = max(h1, h2)
        writer_canvas.setPageSize((out_w, out_h))

        # Draw both pages as images
        writer_canvas.doForm(writer_canvas.beginForm(str(i) + "_left", w1, h1))
        writer_canvas.doForm(writer_canvas.beginForm(str(i) + "_right", w2, h2))

        # Embed pages as raw XObjects
        from reportlab.lib.utils import ImageReader
        from io import BytesIO
        from pypdf import PdfWriter

        def page_as_image(page):
            tmp = BytesIO()
            PdfWriter().add_page(page).write(tmp)
            tmp.seek(0)
            return ImageReader(tmp)

        img1 = page_as_image(p1)
        img2 = page_as_image(p2)

        writer_canvas.drawImage(img1, 0, 0, width=w1, height=h1)
        writer_canvas.drawImage(img2, w1, 0, width=w2, height=h2)

        writer_canvas.showPage()

    writer_canvas.save()


if __name__ == "__main__":
    merge_side_by_side(PDF_LEFT, PDF_RIGHT, OUTPUT)
    print("OK ->", OUTPUT)
```

```python
def convert(text):
    text = text[:-2]
    return float(text)

df['CPO Firm% (RGN)'] = df['CPO Firm% (RGN)'].apply(convert)
```

```python
df.dropna(subset=["DLR Days Supply"], inplace=True)
```

```python
x = df["DLR Days Supply"]
y = df['CPO Firm% (RGN)']
res = stats.pearsonr(x, y)

# Print the results
print(f"Pearson correlation coefficient: {res.statistic}")
print(f"P-value: {res.pvalue}")
```
